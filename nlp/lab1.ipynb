```python
import nltk
from nltk.book import text3
from nltk.book import text1
from nltk.probability import FreqDist

fdist = FreqDist(text3)
```

# TASK1

## words that are hapaxes (unique) and those that occur less than 5 times

```python
hapaxes = fdist.hapaxes()
print(f"Кількість унікальних слів: {len(hapaxes)}")
print(f"Перші 10 унікальних слів: {hapaxes[:10]}")
```

```python
less_than_5_times = [word for word, freq in fdist.items() if freq < 5]
print(f"\nКількість слів, що зустрічаються менше 5 разів: {len(less_than_5_times)}")
print(f"Приклади слів, що зустрічаються менше 5 разів: {less_than_5_times[:10]}")
```

## draw a plot showing mapping from word length to word frequency

```python
import matplotlib.pyplot as plt

word_lengths = [len(word) for word in text3]
fdist_lengths = FreqDist(word_lengths)

plt.figure(figsize=(12, 6))
fdist_lengths.plot()
plt.xlabel('Довжина слова')
plt.ylabel('Кількість слів')
plt.show()
```

## find 10 most frequent words occurring at the end of the sentence

```python
end_words = [text3[i] for i in range(len(text3) - 1) if text3[i+1] == '.']
fdist_end_words = FreqDist(end_words)

print("10 найчастіших слів у кінці речення:")
print(fdist_end_words.most_common(10))
```
